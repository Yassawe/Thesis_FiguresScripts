```python
def plotScatter_vgg(acc, time, labels, savepath):
    # Scatter plot
    plt.figure(figsize=(10, 5), facecolor="white")

    colors = ['cornflowerblue', 'peachpuff', 'palegreen', 'tomato', "purple"]

    plt.scatter(time, acc, s=200, c=colors, edgecolors='black', zorder=3)


    #positions = [(0,12), (30,10), (0, -20), (0,12), (-40,-20)]
    positions = [(0,12), (0,12), (0, -20), (0,12), (0, -20)]

    # Add labels to the points
    for i, label in enumerate(labels):
        plt.annotate(label, (time[i], acc[i]), textcoords="offset points", xytext=positions[i], ha='center', fontsize=14, fontweight='bold')

    # Set axis labels and titles
    plt.xlabel('Time per iteration, ms', fontsize=16)
    plt.ylabel('Test Accuracy, %', fontsize=16)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.grid(color='gray', linestyle='dashed')

    plt.xlim(35, 46)
    plt.ylim(85,95)

    plt.savefig(savepath, dpi=600, bbox_inches="tight")
    
```

```python
def plotScatter_bert(acc, time, labels, savepath):
    # Scatter plot
    plt.figure(figsize=(10, 5), facecolor="white")

    colors = ['cornflowerblue', 'peachpuff', 'palegreen', 'tomato', "purple"]

    plt.scatter(time, acc, s=200, c=colors, edgecolors='black', zorder=3)


    positions = [(0,12), (0,12), (0, -20), (0,12), (10,-20)]
    #positions = [(0,12), (0,12), (0, 12), (0,-25), (-10, 10)]

    # Add labels to the points
    for i, label in enumerate(labels):
        plt.annotate(label, (time[i], acc[i]), textcoords="offset points", xytext=positions[i], ha='center', fontsize=14, fontweight='bold')

    #plt.annotate(text='', xy=(1,1), xytext=(0,0), arrowprops=dict(arrowstyle='<->'))
    
    # Set axis labels and titles
    plt.xlabel('Time per iteration, ms', fontsize=16)
    plt.ylabel('F1 Score', fontsize=16)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.grid(color='gray', linestyle='dashed')

    plt.xlim(300, 330)
    plt.ylim(83,93)

    plt.savefig(savepath, dpi=600, bbox_inches="tight")
    
```
